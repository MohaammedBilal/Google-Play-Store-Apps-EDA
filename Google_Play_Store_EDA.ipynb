## **6.8 Size Column**
Contains the size of the apps

Our Goal:

1.   Imputing Missing Values
2.   Changing the data type



```python
df["Size"].isnull().sum()
```

```python
df["Size"].unique()
```



There are sizes in " M " and " k ". We will convert all values in " M " first and then remove word " M " from values.

```python
# Replace 'Varies with device' with NaN
df['Size'] = df['Size'].replace('Varies with device', np.nan)

# Function to convert size to MB
def convert_size_to_mb(size):
    if pd.isna(size):
        return np.nan
    # Remove commas before conversion
    size = str(size).replace(',', '')
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    return np.nan # Handle any other unexpected formats

# Apply the conversion function
df['Size'] = df['Size'].apply(convert_size_to_mb)

# Check the data type and missing values
print(f"Data type of 'Size' after conversion: {df['Size'].dtype}")
print(f"Missing values in 'Size' after conversion: {df['Size'].isnull().sum()}")
display(df[['Size']].head())
```

```python
df.describe()
```

We can clearly see that data type of Size column changes

```python
df["Size"].unique()
```

Now Imputing Missing Values of Size Column

```python
# Impute missing values in 'Size' with the median size
median_size = df['Size'].median()
df['Size'].fillna(median_size, inplace=True)
```

## **6.9 Installs and Minimum Installs Columns**

```python
df["Installs"].unique()
```

*   In Install column there is " + " sign which indicates that it is categorical column and describing the approximate range. It can be used to give names using categories.
*   Maximum Installs column is more sufficient for analysis as it is providing the number of installs which we will consider in our analysis.



```python
# Inspecting the rows where installs values are missing
df[df["Installs"].isnull()].head(5)
```

#### Checking wether the missing values in installs column are all those which have missing values in minimum installs


```python
df[df["Installs"].isnull()]["Minimum Installs"].isnull().sum()
```


It means all in both columns same row values are missing.

###**Our Approach**

We will keep installs column as it is with + sign. for imputing the installs and minimum installs col , we will check that if max installs is greater than 10 then impute installs as 10+ if more than 100 than impute installs as 100+ if greater than 1000 impute installs as 1,000+ if greater than 5000 impute as 5,000+ and also give same floating value to minimum installs


```python
# Impute missing values in 'Installs' and 'Minimum Installs' based on 'Maximum Installs'
for index, row in df[df['Installs'].isnull()].iterrows():
    max_installs = row['Maximum Installs']
    if max_installs > 5000:
        df.loc[index, 'Installs'] = '5,000+'
        df.loc[index, 'Minimum Installs'] = 5000.0
    elif max_installs > 1000:
        df.loc[index, 'Installs'] = '1,000+'
        df.loc[index, 'Minimum Installs'] = 1000.0
    elif max_installs > 100:
        df.loc[index, 'Installs'] = '100+'
        df.loc[index, 'Minimum Installs'] = 100.0
    elif max_installs > 10:
        df.loc[index, 'Installs'] = '10+'
        df.loc[index, 'Minimum Installs'] = 10.0
    else:
        # For cases where max_installs is 10 or less, you might choose a different
        # category or keep them as NaN depending on your desired granularity.
        # Here, I'll impute as '0+' and 0.0 for simplicity based on the logic provided.
        df.loc[index, 'Installs'] = '0+'
        df.loc[index, 'Minimum Installs'] = 0.0

# Verify the imputation
print(f"Missing values in 'Installs' after imputation: {df['Installs'].isnull().sum()}")
print(f"Missing values in 'Minimum Installs' after imputation: {df['Minimum Installs'].isnull().sum()}")
```
